# Introducción a las pruebas de hipotesis

Las pruebas de hipótesis son evaluaciones que se hacen con datos para ver si una idea que tenemos sobre una población es correcta o si podría ser solo una casualidad. Por ejemplo, en un ensayo clínico para un nuevo medicamento para el resfriado, podríamos preguntarnos si el medicamento realmente ayuda a la gente a recuperarse más rápido o si los resultados que vemos podrían deberse al azar.

Para averiguarlo, un experto en datos realiza una prueba que nos ayuda a calcular si lo que estamos viendo podría ser algo poco común o si es algo que podría ocurrir por pura casualidad. Si los resultados son muy poco comunes y es poco probable que sucedan por azar, entonces decimos que son estadísticamente significativos. Esto nos ayuda a decidir si el medicamento realmente funciona o si necesitamos hacer más investigaciones.

# Importancia en la toma de decisiones.

Una razón para hacer pruebas de hipótesis es ayudar en la toma de decisiones. Básicamente, cuando tomamos una decisión basada en la prueba, esta decisión se basa en lo que nos dice la estadística. Si decimos que no creemos en algo, generalmente es porque creemos en algo diferente. Si no rechazamos la idea inicial, entonces estamos diciendo que no creemos en la alternativa. Sin embargo, a veces también podemos decidir recopilar más información en lugar de tomar una decisión definitiva.

Pero aquí hay algo importante que debemos recordar: el resultado de la prueba estadística no es la única cosa en la que basamos nuestra decisión final. No debemos pensar que la decisión estadística es la última palabra, sino que debemos considerar toda la información relevante que tengamos como experimentadores. Así que, en resumen, cuando tomamos decisiones basadas en pruebas de hipótesis, no solo miramos la estadística, sino que también consideramos otros datos importantes. Esto es lo que vamos a explorar más a fondo al estudiar algunas pruebas de hipótesis específicas comunes.

# Elementos de una prueba de hipótesis

## Hipótesis Nula (H0) y Alternativa (Ha)

Una prueba de hipótesis forma parte de la teoría de la decisión y examina dos hipótesis opuestas sobre una muestra poblacional: la hipótesis nula y la hipótesis alternativa. La hipótesis alternativa (Ha) es el enunciado que deseamos concluir como verdadero conforme a la evidencia de los datos de la muestra. Mientras que la hipótesis nula (H0) es el enunciado que se probará y buscará rechazar, por lo general es un enunciado de que "no hay efecto" o "no hay diferencia".

En general H0, se establece con el propósito expreso de ser rechazada. Si no se rechaza, esto no necesariamente significa que es verdadera, se dirá que los datos sobre los cuales se basa la prueba no proporcionan evidencia suficiente que cause el rechazo. Por el contrario, si se rechaza se concluye que los datos disponibles no son compatibles con la H0, pero sirven como apoyo a alguna otra hipótesis. Rechazarla entonces, sugiere que la hipótesis alternativa puede ser verdadera.

## Valor alpha (α)

El valor alfa (α) es una medida de significancia que establecemos antes de realizar una prueba, similar a un margen de error que toleramos en nuestras conclusiones. Por ejemplo, al configurar α en 0.05, estamos dispuestos a aceptar un margen de error del 5% en nuestras conclusiones. Este valor alfa actúa como un umbral que nos ayuda a decidir cuán extremos deben ser los resultados para rechazar la hipótesis nula (H0).

Este concepto está relacionado con el intervalo de confianza (CI). El valor alfa influye en el nivel de confianza que tenemos en nuestras conclusiones. Al fijar α en 0.05, creamos un intervalo de confianza del 95%. Esto significa que estamos dispuestos a aceptar un margen de error del 5% en nuestras conclusiones y que confiamos en un 95% en la validez de nuestras conclusiones.

Dado que rechazar una H0 verdadera sería un error, parece razonable que debamos minimizar la probabilidad de cometerlo, y es precisamente lo que hacemos. Se elige un valor pequeño de α para reducir al mínimo la probabilidad de rechazar una H0 verdadera. Los valores de α que se encuentran con más frecuencia son 0.01, 0.05 y 0.10. La probabilidad de cometer un error al no rechazar una H0 verdadera generalmente se establece en un 95%, aunque en algunos casos puede ser del 90% o del 99%. Esto se conoce como el nivel de confianza y refleja cuánto confiamos en la validez de nuestras conclusiones.

##  Estadístico de prueba
Un estadístico de prueba es una variable aleatoria que se calcula a partir de datos de muestra y se utiliza en una prueba de hipótesis. Puede utilizar los estadísticos de prueba para determinar si puede rechazar la hipótesis nula. El estadístico de prueba compara sus datos con lo que se espera bajo la hipótesis nula. El estadístico de prueba se utiliza para calcular el valor p.

Las diferentes pruebas de hipótesis utilizan diferentes estadísticos de prueba según el modelo de probabilidad asumido en la hipótesis nula. Las pruebas comunes y sus respectivos estadísticos de prueba incluyen:


| Prueba de hipótesis | Estadístico de prueba |
|---------------------|-----------------------|
| Prueba Z            | Estadístico Z         |
| Pruebas t           | Estadístico t         |
| ANOVA               | Estadístico F         |



## Valor p (p-value)

El valor p, o p-value, te dice qué tan probable es que tus datos hayan ocurrido bajo la hipótesis nula. Hace esto calculando la probabilidad de tu estadístico de prueba, que es el número calculado por una prueba estadística utilizando tus datos.

El valor p te dice con qué frecuencia esperarías ver un estadístico de prueba tan extrema o más extrema que la calculada por tu prueba estadística si la hipótesis nula de esa prueba fuera cierta. El valor p se hace más pequeño a medida que el estadístico de prueba calculada a partir de tus datos se aleja más del rango de estadísticos de prueba predichas por la hipótesis nula.

El valor p es una proporción: si tu valor p es 0.05, eso significa que el 5% del tiempo verías un estadístico de prueba al menos tan extrema como la que encontraste si la hipótesis nula fuera cierta. Para tomar decisiones en una prueba de hipótesis, comparamos el valor p con el nivel de significancia (alfa) que previamente establecimos. Si el valor p es menor que alfa, rechazamos la hipótesis nula; de lo contrario, no la rechazamos.


  * Si el valor p es menor que α, entonces rechazamos (reject)la hipótesis nula.
  * Si el valor p es mayor o igual que α, entonces no rechazamos (fail to reject) la hipótesis nula.

## Tipos de error

Ninguna prueba de hipótesis garantiza una certeza absoluta. Dado que las pruebas se basan en probabilidades, siempre existe la posibilidad de llegar a una conclusión incorrecta. Al realizar una prueba de hipótesis, se enfrenta a dos tipos de errores: el Error tipo I (falso positivo) y el Error tipo II (falso negativo). La probabilidad de cometer estos errores está relacionada con el nivel de significancia que elige. Por lo tanto, es esencial evaluar cuál de estos errores tiene consecuencias más graves para su situación antes de definir los riesgos.

### __Error tipo I (False Positive)__
Este error ocurre cuando rechaza la hipótesis nula cuando en realidad es cierta, lo que se conoce como un falso positivo. La probabilidad de cometer un error tipo I se representa como α y refleja el nivel de significancia que establece para su prueba. Por ejemplo, si configura α en 0.05, está dispuesto a aceptar un 5% de probabilidad de equivocarse al rechazar la hipótesis nula. Para disminuir este riesgo, puede optar por un valor más bajo para α.

### __Error de tipo II (False Negative)__
Este error se produce cuando la hipótesis nula es falsa y usted no la rechaza, lo que se llama un falso negativo. La probabilidad de cometer un error de tipo II se denota como β y depende de la potencia de la prueba. Puede reducir el riesgo de cometer un error de tipo II asegurándose de que la prueba tenga suficiente potencia. Esto implica utilizar un tamaño de muestra lo suficientemente grande como para detectar diferencias significativas cuando realmente existen.

__El Dilema de los Errores en las Pruebas de Hipótesis__

Para reducir el error tipo I, es decir, la probabilidad de cometer un falso positivo en una prueba de hipótesis, puedes optar por un valor de nivel de significancia (alpha, α) más bajo. Por ejemplo, si configuras α en 0.01 en lugar de 0.05, estás siendo más estricto en cuanto a la evidencia requerida para rechazar la hipótesis nula. Esto significa que serás más cauteloso al rechazar la hipótesis nula y, por lo tanto, reducirás la probabilidad de cometer un error tipo I.

Sin embargo, aquí está el trade-off (compensación): cuando eliges un valor de α más bajo para reducir el error tipo I, aumenta la probabilidad de cometer un error tipo II, que es la probabilidad de no rechazar la hipótesis nula cuando en realidad es falsa (un falso negativo). Esto sucede porque estás siendo más exigente en la evidencia requerida para rechazar la hipótesis nula, lo que hace que sea más difícil detectar efectos reales en tus datos.

## Potencia Estadística

La potencia estadística se refiere a la capacidad de una prueba para detectar efectos o diferencias reales en los datos. En otras palabras, es la probabilidad de que la prueba sea capaz de identificar una diferencia significativa si esta realmente existe en la población. Esta medida está relacionada de manera inversa con el error de tipo II (β), que es la probabilidad de no rechazar la hipótesis nula cuando en realidad es falsa, es decir, un falso negativo.

Cuando hablamos de la potencia estadística, estamos hablando de cuán efectiva es una prueba para evitar cometer un error de tipo II. Cuanto mayor sea la potencia estadística, menor será la probabilidad de cometer un falso negativo, lo que significa que la prueba tiene una mayor capacidad para detectar efectos reales si realmente están presentes en los datos. Por lo tanto, la potencia estadística y el error de tipo II están inversamente relacionados: a medida que la potencia estadística aumenta, la probabilidad de cometer un error de tipo II disminuye.

Para mejorar la potencia estadística de una prueba:

* Aumenta el tamaño de la muestra.
* Considera un nivel de significancia ligeramente más alto.
* Mejora la precisión de las mediciones.
* Escoge pruebas estadísticas adecuadas.
* Extiende el período de observación.
* Diseña experimentos de manera eficiente.
* Realiza un análisis de potencia antes del estudio.

# Pasos generales para realizar una prueba de hipótesis:
Para llevar a cabo una prueba de hipótesis, se siguen los siguientes pasos:

* Establece las Hipótesis: Comienza definiendo la hipótesis nula (H0) y la hipótesis alternativa (Ha). La H0 es la afirmación inicial que deseas poner a prueba, mientras que la Ha es la afirmación contraria o alternativa que estás considerando.

* Selecciona el Nivel de Significancia (Alpha): Elige un nivel de significancia (α) que represente el margen de error que estás dispuesto a aceptar. Comúnmente se utiliza α = 0.05, lo que significa que estás dispuesto a aceptar un 5% de probabilidad de cometer un error de tipo I.

* Calcula el Estadístico de Prueba: Utiliza una prueba estadística apropiada (como una prueba t o una prueba z) para calcular un valor estadístico basado en tus datos.

* Encuentra el Valor p (p-value): El valor p es una medida que indica cuán probable es que los resultados observados sean consistentes con la hipótesis nula. Cuanto menor sea el valor p, más fuerte es la evidencia en contra de la H0.

* Toma una Decisión: Compara el valor p calculado con tu nivel de significancia (α). Si el valor p es menor que α, rechazas la hipótesis nula (H0). Si el valor p es mayor que α, no rechazas la H0.


## Pruebas de Una Cola (One-Tailed Tests )y Dos Colas (Two-Tailed Tests)

Un __"One-tailed test"__ (prueba de una cola) ocurre cuando la hipótesis alternativa afirma que el valor real de un parámetro de población es o bien menor o mayor que el valor en la hipótesis nula.

Esta prueba puede ser "de cola izquierda" o "de cola derecha". La prueba de cola izquierda ocurre cuando la hipótesis alternativa sostiene que el valor real del parámetro es menor que el valor en la hipótesis nula. En cambio, la prueba de cola derecha ocurre cuando la hipótesis alternativa sostiene que el valor real del parámetro es mayor que el valor en la hipótesis nula.

Por otro lado, un __"Two-tailed test"__ (prueba de dos colas) se da cuando la hipótesis alternativa afirma que el valor real del parámetro no es igual al valor en la hipótesis nula.

Tomemos como ejemplo una prueba en la que la hipótesis nula establece que el peso promedio de una población de pingüinos es igual a 30 libras.

#  Pruebas de Hipótesis de Una Muestra (One Sample) y Dos Muestras (Two Sample)

## Pruebas de Hipótesis de Una Muestra (One Sample):

La prueba de hipótesis de una muestra se utiliza para responder a la pregunta de si una muestra de datos es representativa de una población más grande. Básicamente, se trata de verificar si los datos de la muestra son similares o diferentes de lo que se esperaría en una situación normal. Para hacer esto, se compara la muestra con una suposición inicial llamada hipótesis nula, y se busca determinar si hay evidencia suficiente para rechazar esa suposición.

__Ejemplo 1 : (Precio Promedio del Refresco):__ Supongamos que tenemos una base de datos que contiene el precio promedio de una marca de refresco en una región. Queremos determinar si el precio promedio actual es significativamente diferente del precio objetivo de $2.50 por botella (hipótesis nula H0).
La hipótesis alternativa (Ha) sería que el precio promedio es diferente de \$2.50 por botella. Utilizaríamos una prueba de hipótesis de una muestra para evaluar si existe una diferencia significativa en el precio promedio en comparación con el valor de referencia.

__Ejemplo 2 (Preferencia de Marca de Refresco):__ Ahora, cambiemos nuestro enfoque a las preferencias de marca de refresco. Supongamos que deseamos determinar si al menos el 70% de las personas prefieren una marca específica (hipótesis nula H0). La hipótesis alternativa (Ha) sería que menos del 70% de las personas prefieren esta marca. Usaríamos una prueba de hipótesis de una muestra para evaluar si la proporción de personas que prefieren esta marca es significativamente menor al 70%.

## Pruebas de Hipótesis de Dos Muestras (Two Sample):

La prueba de hipótesis de dos muestras se utiliza para comparar dos grupos de datos independientes y determinar si existen diferencias significativas entre ellos. A continuación, se presentan ejemplos utilizando variables continuas y variables discretas en el contexto de las preferencias de marca de refresco.

__Ejemplo 1 (Precio Promedio del Refresco entre Dos Marcas):__ Imaginemos que tenemos datos de precios promedio de dos marcas de refresco diferentes en una región. Queremos comparar si hay una diferencia significativa en los precios promedio entre estas dos marcas. La hipótesis nula (H0) podría afirmar que no hay diferencia significativa en los precios promedio entre las dos marcas, mientras que la hipótesis alternativa (Ha) sugeriría que existe una diferencia significativa. En este caso, usaríamos una prueba de hipótesis de dos muestras para comparar los precios promedio entre las dos marcas.

__Ejemplo 2 (Preferencia de Marca de Refresco entre Dos Grupos de Edad):__ Supongamos que queremos comparar la preferencia de marca de refresco entre dos grupos de edad: Grupo A (18-30 años) y Grupo B (31-50 años). La hipótesis nula (H0) podría afirmar que no hay diferencia significativa en la preferencia de marca entre los dos grupos de edad, mientras que la hipótesis alternativa (Ha) sugeriría que existe una diferencia significativa. Utilizaríamos una prueba de hipótesis de dos muestras para comparar las preferencias de marca entre los grupos de edad.

# A/B Testing y Pruebas de Hipótesis

El A/B testing, también conocido como prueba A/B, es una técnica ampliamente utilizada en el análisis de datos y el marketing digital. A menudo se considera una forma de prueba de hipótesis de dos muestras. En un A/B test, se comparan dos grupos, el grupo de control (A) y el grupo variable (B), para evaluar si existe una diferencia significativa en función de una métrica específica, como tasas de conversión, tasas de clics o tiempo en el sitio web.

Los pasos típicos en un A/B test incluyen:

* Grupo de Control vs. Grupo Variable: Se crean dos grupos, uno expuesto a la versión actual o de control y otro expuesto a una versión modificada o variable del elemento que se está probando.

* Definición de Hipótesis: Se establecen dos hipótesis: la hipótesis nula (H0), que sostiene que no hay diferencia significativa entre los grupos, y la hipótesis alternativa (Ha), que sugiere lo contrario.

* Recopilación de Datos: Se recopilan datos mientras ambos grupos interactúan con las diferentes versiones. Se registran métricas clave.

* Análisis Estadístico: Se aplica un análisis estadístico, a menudo utilizando pruebas de hipótesis de dos muestras, para determinar si la diferencia observada es estadísticamente significativa.

* Toma de Decisiones: Basándose en los resultados del análisis, se toma una decisión informada sobre si implementar o no la versión modificada.

El A/B testing es una herramienta poderosa para optimizar estrategias de marketing y diseño de productos en línea al permitir decisiones basadas en datos. Es una aplicación práctica de las pruebas de hipótesis que puede ayudar a mejorar el rendimiento y la eficacia en diversas áreas.
